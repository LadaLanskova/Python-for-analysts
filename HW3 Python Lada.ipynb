```python
df = pd.read_csv(r'C:\Users\Geralt\tmp\.env\Scripts\kc_house_data.csv', encoding = 'cp1251')
df.head()
```

### 1.1.
Создайте новый признак price_per_sq_lot, который будет содержать стоимость за 1 кв. метр общей площади.

```python
df['price_per_sq_lot'] = df['price'] / df['sqft_lot']
df.head()
```

### 1.2 
Создать новый признак delta_renovated, который будет содержать разницу в годах между годом реновацией дома и годом постройки дома. Если реновации дома не было, то в новом признаке поставьте 0

```python
df['delta_renovated'] = df['yr_renovated'] - df['yr_built']
df.loc[df['delta_renovated'] <0, 'delta_renovated'] = 0
df
```

### 1.3 
Создайте признаки года продажи, месяца продажи

```python
df['deal_year'] = df['date'].apply(lambda x: str(x)[0:4])
df['deal_month'] = df['date'].apply(lambda x: str(x)[4:6])
df
```

### 1.4 
Удалите признаки date, zipcode, lat, long

```python
df.drop(columns=['date', 'zipcode', 'lat', 'long'], inplace=True)
df
```

## Задача 3.

### 3.1 
Найдите среднюю стоимость домов в зависимости от количества спален
Отсортируйте от меньшей стоимости к большей

```python
df.groupby('bedrooms').agg({'price': 'mean'}).sort_values('price')
```

### 3.2 
Найдите минимальную, среднюю и максимальную стоимости домов в зависимости от состояния дома

```python
# Первый вариент через groupby
condition = df[['price', 'condition']].groupby('condition')
condition.agg({'price': ['min', 'mean', 'max']})
```

```python
# Второй вариант через pivot table
df.pivot_table(index = 'condition', values = 'price', aggfunc = ['min', 'mean', 'max'])
```

### 3.3 
Постройте таблицу с подсчетом количества домов в данных в зависимости от вида на набережную и оценкой вида

### 3.4 
Каких домов в зависимости от этажности и количества спален больше?

```python
df_c = df[['id', 'floors', 'bedrooms']].groupby(['floors', 'bedrooms']).count()
df_c[df_c['id'] == df_c['id'].max()]
```
